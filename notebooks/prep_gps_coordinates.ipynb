```python
import time
import numpy as np
import pandas as pd
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
```

```python
def get_gps_coordinates(city):
    geolocator = Nominatim(user_agent="specify_your_app_name_here")
    location = geolocator.geocode(city, timeout=10)
    return (location.latitude, location.longitude)
```

```python
def compute_distance_from_names(city_1, city_2):
    coords_1 = get_gps_coordinates(city_1)
    coords_2 = get_gps_coordinates(city_2)
    return geodesic(coords_1, coords_2).miles
```

```python
def compute_distance_from_coords(coords_1, coords_2):
    return geodesic(coords_1, coords_2).miles
```
